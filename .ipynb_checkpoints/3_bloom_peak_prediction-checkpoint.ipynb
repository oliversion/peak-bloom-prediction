```python
import math
from datetime import datetime, timedelta
import numpy as np
import pandas as pd
from random import sample
import matplotlib.pyplot as plt
```

```python
df_monthly_imp = pd.read_csv('data_ext/all_locations_weather_monthly_imp.csv')
df_monthly_imp.bloom_date = df_monthly_imp.bloom_date.apply(lambda x: datetime.strptime(x, '%Y-%m-%d').date())
df_monthly_imp.head(2)
```

## Use all available data to train regressors

```python
X_train = df_monthly_imp.drop(["location", "bloom_date", "bloom_doy"], axis=1) 
y_train = df_monthly_imp["bloom_doy"]
```

Using validation set, it was observed that ensamble of GradientBoostingRegressor, CatBoostRegressor, XGBRegressor outperform other models. Optimal parameters of each regressor was found using grid search. Thus, this ensamble model is used to predict the peak bloom date for each year of the next decade (2023 to 2032) at for locations.

```python
from sklearn.ensemble import GradientBoostingRegressor
from xgboost.sklearn import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import accuracy_score, classification_report, mean_absolute_error,r2_score,mean_squared_error
```

```python
# train regressors
gb_reg = GradientBoostingRegressor(random_state=0, learning_rate=0.05, max_depth=5, n_estimators=400)
gb_reg.fit(X_train, y_train)

cat_reg = CatBoostRegressor()
cat_reg.fit(X_train, y_train)

xgb_reg = XGBRegressor(random_state=0, learning_rate=0.1, n_estimators=300)
xgb_reg.fit(X_train, y_train)
```

```python
features = list(X_train.columns)
sorted_feature_importance = cat_reg.feature_importances_.argsort()
plt.figure(figsize=(8,9))
plt.barh([features[i] for i in sorted_feature_importance], 
        cat_reg.feature_importances_[sorted_feature_importance], 
        color='turquoise')
plt.xlabel("CatBoost Feature Importance")
```

## Make predictions for the next decade using location and collected meteorological data

```python
test_kyoto = pd.read_csv('data_ext/kyoto_test.csv') 
test_liestal = pd.read_csv('data_ext/liestal_test.csv') 
test_washingtondc = pd.read_csv('data_ext/washingtondc_test.csv') 
test_vancouver = pd.read_csv('data_ext/vancouver_test.csv') 
```

```python
y_pred_kyoto = np.mean(np.array([gb_reg.predict(test_kyoto), cat_reg.predict(test_kyoto), 
                                 xgb_reg.predict(test_kyoto)]), axis=0)
y_pred_kyoto
```

```python
y_pred_test_liestal = np.mean(np.array([gb_reg.predict(test_liestal), cat_reg.predict(test_liestal), 
                                 xgb_reg.predict(test_liestal)]), axis=0)
y_pred_test_liestal
```

```python
y_pred_washingtondc = np.mean(np.array([gb_reg.predict(test_washingtondc), cat_reg.predict(test_washingtondc), 
                                 xgb_reg.predict(test_washingtondc)]), axis=0)
y_pred_washingtondc
```

```python
y_pred_vancouver = np.mean(np.array([gb_reg.predict(test_vancouver), cat_reg.predict(test_vancouver), 
                                 xgb_reg.predict(test_vancouver)]), axis=0)
y_pred_vancouver
```

```python
df_pred = pd.DataFrame(columns=['year','kyoto', 'liestal','washingtondc', 'vancouver'])
df_pred.year = test_kyoto.year
df_pred.kyoto = y_pred_kyoto
df_pred.liestal = y_pred_test_liestal
df_pred.washingtondc = y_pred_washingtondc
df_pred.vancouver = y_pred_vancouver
df_pred
```

```python
df_pred = df_pred.round().astype('int32')
df_pred
```

```python
df_pred.to_csv('data_ext/prediction.csv', index=False) 
```
